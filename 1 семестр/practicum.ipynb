# 1
 Реализовать программу, с которой можно играть в логическую игру «Быки и коровы».
 Программа загадывает число, пользователь вводит очередной вариант отгадываемого числа, программа возвращает количество быков и коров и в случае выигрыша игрока сообщает о победе и завершается. Сама программа НЕ ходит, т.е. не пытается отгадать число загаданное игроком.
 
Взаимодействие с программой производится через консоль, при запросе данных от пользователя программа сообщает, что ожидает от пользователя и проверяет корректность ввода.

```python
import random



def num_validation(num): 
    """ Проверяем, все ли цифры в чиле отличны друг от друга """
    num_as_list = list(str(num))
    if len(set(num_as_list)) == 4:
        return True
    else:
        return False

def count_bulls_cows(est_number, num_to_guess):
    """ Находим количество коров и быков в числе """
    cows_list, bulls_list = [], []    # индексы коров и быков
    input_number_list = list(str(est_number)) # число вводимое пользователем (в виде списка)
    myst_number_list = list(str(num_to_guess)) # число, которое нужно угадать
    
    for i in range(len(input_number_list)):
        if myst_number_list[i] == input_number_list[i]:
            bulls_list.append(i)
        elif input_number_list[i] in myst_number_list:
            cows_list.append(i)
    print(f"Найдено {len(cows_list)} коров и {len(bulls_list)} быков")
    return cows_list, bulls_list


def game():
    count = 0  # счетчик ходов
    myst_num = 0 
    bulls_list = []  # число быков
    cows_list = []   # число коров
    

    while not num_validation(myst_num):   # генерируем случайное число, которе будет иметь 4 различные цифры в составе
        myst_num = random.randint(1000, 9000)

    while len(bulls_list) != 4:
        input_num = input("Введите четырёхзначное число: ")
        count += 1
        bulls_list, cows_list = count_bulls_cows(input_num, myst_num) # проверяем каждое введенное число на соответствие загаданному числу

        if len(bulls_list) == 4:
            print(f"Победа! Вы угадали число {myst_num} за {count} ходов!")
            break

game()
```

Результат выполнения операции вернуть в виде текстового представления числа. Пример calc("двадцать пять плюс тринадцать") -> "тридцать восемь"

Оформить калькулятор в виде функции, которая принимает на вход строку и возвращает строку.

```python
def calculate(str_exp):

    plus = lambda x, y: x + y 
    minus = lambda x, y: x - y
    mult = lambda x, y: x * y

# данные для перевода числа из строкового представления в числовое
    data2calc = {'девяносто': 90, 'восемьдесят': 80, 'семьдесят': 70, 'шестьдесят': 60, 'пятьдесят': 50,
                     'сорок': 40, 'тридцать': 30, 'двадцать': 20, 'девятнадцать': 19, 'восемнадцать': 18,
                     'семнадцать': 17, 'шестнадцать': 16, 'пятнадцать': 15, 'четырнадцать': 14, 'тринадцать': 13,
                     'двенадцать': 12, 'одиннадцать': 11, 'десять': 10, 'девять': 9, 'восемь': 8, 'семь': 7,
                     'шесть': 6, 'пять': 5, 'четыре': 4, 'три': 3, 'два': 2, 'один': 1, 'ноль': 0,
                     'плюс': plus, 'минус': minus, 'умножить': mult}
# данные для перевода числа из числового представления в строковое
    output_data = {9000: 'девять тысяч', 8000: 'восемь тысяч', 7000: 'семь тысяч', 6000: 'шесть тысяч',
                  5000: 'пять тысяч', 4000: 'четыре тысячи', 3000: 'три тысячи', 2000: 'две тысячи',
                  1000: 'одна тысяча', 900: 'девятьсот', 800: 'восемьсот', 700: 'семьсот', 600: 'шестьсот',
                  500: 'пятьсот', 400: 'четыреста', 300: 'триста', 200: 'двести', 100: 'сто',
                  90: 'девяноста', 80: 'восемьдесят', 70: 'семдесят', 60: 'шестьдесят', 50: 'пятьдесят',
                  40: 'сорок', 30: 'тридцать', 20: 'двадцать', 19: 'девятнадцать', 18: 'восемнадцать',
                  17: 'семнадцать', 16: 'шестнадцать', 15: 'пятнадцать', 14: 'четырнадцать', 13: 'тринадцать',
                  12: 'двенадцать', 11: 'одиннадцать', 10: 'десять', 9: 'девять', 8: 'восемь', 7: 'семь',
                  6: 'шесть', 5: 'пять', 4: 'четыре', 3: 'три', 2: 'два', 1: 'один', 0: 'ноль'}

    str2num = [] 
    oper_symbol = 0

    for i, item in enumerate(str_exp.split()):
            str2num.append(data2calc[item])
            if isinstance(data2calc[item], type(lambda: None)):
                oper_symbol = i # получаем индекс "операции" в введенном выражении (список str2num)


    res_num = str2num[oper_symbol](sum(str2num[:oper_symbol]),
                                            sum(str2num[oper_symbol + 1:])) # вызываем лямбда функцию по индесу из списка и передаем ей 
                                                                            # суммы эл-тов срезов списка
    num2str = []

    if res_num < 0:
        num2str.append('минус')
        res_num = abs(res_num)
    elif res_num == 0:
        return print('ноль')

    for i in reversed(range(1, 4)):
        j = res_num // 10 ** i * 10 ** i # определяем число тысяч, сотен, десятков
        if j == 0:
            continue
        elif (j < 20) & (i == 1):
            num2str.append(output_data[res_num]) 
            break
        else:
            num2str.append(output_data[j])
        res_num -= j
        #if res_num < 10:                         # либо проверяем "оставшееся" число в конце цикла и если оно меньше 10 добавляем его в строку
            #num2str.append(output_data[res_num])
    else:                                         # либо используем else, который выполняется если цикл завершился "нормально" (не break)
        num2str.append(output_data[res_num])
    return print(' '.join(num2str))


exp_1 = "восемьдесят три плюс тридцать шесть" # сложение 
exp_2 = "одиннадцать умножить двадцать пять" # умножение
exp_3 = "тридцать три минус шестьдесят пять" # вычитание/ результат - отрицательное число
exp_4 = "сорок восемь умножить двадцать шесть" # результат - четырехзначное число
exp_5 = "ноль умножить ноль" # результат - ноль
exp_6 = "двадцать восемь умножить сорок один" # 1148
exp_7 = "девяносто девять умножить девяносто девять" # максимальное число для калькулятора 9801

calculate(exp_1)
calculate(exp_2)
calculate(exp_3)
calculate(exp_4)
calculate(exp_5)
calculate(exp_6)
calculate(exp_7)

```
